# South African Healthcare AI System

## Objective
Multi-agent AI system to assist triage, medical guidance, and follow-up care using Groq + LangGraph + FAISS.

## Architecture

Patient Query
      |
      v
Supervisor Agent (LangGraph Router)
      |
      +--> Triage Agent
      +--> Medical RAG Agent (FAISS)
      +--> Follow-Up Agent
      +--> Translation Agent
      |
      v
Final Response

## Tech Stack
- Groq (Llama 3.3 70B)
- LangGraph
- FAISS Vector Database
- Sentence Transformers
- Python (Colab)

“This system is designed as an agentic workflow, not a chatbot, enabling dynamic routing and modular clinical decision support.”

```python
from typing import TypedDict
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage
import os
```

```python
class HealthcareState(TypedDict):
    patient_query: str
    triage_result: str
    medical_response: str
    followup_plan: str
```

```python
def triage_agent(state: HealthcareState):

    query = state["patient_query"]

    prompt = f"""
    You are a healthcare triage assistant.

    Classify the patient into:

    - EMERGENCY
    - URGENT
    - ROUTINE

    Symptoms:

    {query}

    Return:
    Risk Level
    Explanation
    """

    response = llm.invoke(prompt)

    return {
        "triage_result": response.content
    }
```

```python
def medical_agent(state: HealthcareState):

    query = state["patient_query"]

    prompt = f"""
    You are a medical assistant for South Africa.

    Patient query:

    {query}

    Provide:

    - Possible causes
    - Recommended actions
    - Warning signs

    Do NOT diagnose.
    """

    response = llm.invoke(prompt)

    return {
        "medical_response": response.content
    }
```

```python
def followup_agent(state: HealthcareState):

    prompt = f"""
    Based on this triage result:

    {state['triage_result']}

    Create:

    - Follow-up instructions
    - Medication reminder suggestions
    - Next appointment recommendation
    """

    response = llm.invoke(prompt)

    return {
        "followup_plan": response.content
    }
```

```python
def route_patient(state):

    triage = state["triage_result"].upper()

    if "EMERGENCY" in triage:
        return "emergency"

    return "normal"
```

```python
def emergency_agent(state: HealthcareState):

    return {
        "medical_response":
        """
        EMERGENCY DETECTED

        Seek immediate medical attention.
        Call emergency services.
        Visit nearest hospital immediately.
        """
    }
```

```python
workflow = StateGraph(HealthcareState)

workflow.add_node("triage", triage_agent)
workflow.add_node("medical", medical_agent)
workflow.add_node("followup", followup_agent)
workflow.add_node("emergency", emergency_agent)

workflow.set_entry_point("triage")

workflow.add_conditional_edges(
    "triage",
    route_patient,
    {
        "emergency": "emergency",
        "normal": "medical"
    }
)

workflow.add_edge("medical", "followup")

workflow.add_edge("followup", END)
workflow.add_edge("emergency", END)

app = workflow.compile()
```

```python
def memory_agent(state):
    history = state.get("messages", [])

    history.append({
        "role": "user",
        "content": state["patient_query"]
    })

    return {"messages": history}
```

```python
def calculator_tool(expression: str):
    try:
        return str(eval(expression))
    except:
        return "Invalid expression"
```

```python
def symptom_checker(symptoms: str):

    symptoms = symptoms.lower()

    if "chest pain" in symptoms and "breathing" in symptoms:
        return "HIGH RISK: possible cardiac or respiratory emergency"

    if "fever" in symptoms and "cough" in symptoms:
        return "Possible infection (TB / flu / pneumonia)"

    return "Low risk but monitor symptoms"
```

```python
def tool_router(state):

    query = state["patient_query"]

    if "calculate" in query:
        return "calculator"

    if "symptom" in query:
        return "symptom_checker"

    return "medical"
```

```python
def structured_medical_output(state):

    prompt = f"""
    Convert this into a medical report JSON:

    {state['medical_response']}

    Return ONLY JSON with:

    {{
      "risk_level": "",
      "possible_conditions": [],
      "recommended_actions": [],
      "follow_up_days": number,
      "emergency": true/false
    }}
    """

    response = llm.invoke(prompt)

    return {"final_response": response.content}
```

```python
def hallucination_checker(state):

    prompt = f"""
    You are a medical safety validator.

    Check if this response is medically unsafe or hallucinated:

    RESPONSE:
    {state['medical_response']}

    Rules:
    - No fake drugs
    - No false diagnosis
    - Must align with general medical guidelines

    Return:
    SAFE or UNSAFE + reason
    """

    result = llm.invoke(prompt)

    return {"evaluation": result.content}
```

```python
def safety_router(state):

    if "UNSAFE" in state.get("evaluation", ""):
        return "retriage"

    return "final"
```
